```python
!pip install nflows
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
from nflows.flows.base import Flow
from nflows.transforms import CompositeTransform, ReversePermutation, PiecewiseRationalQuadraticCouplingTransform
from nflows.distributions import StandardNormal
from nflows.nn.nets import ResidualNet
from scipy.stats import gaussian_kde

# 1. Generate Data
# Gaussian Mixture
np.random.seed(42)
mean1 = [0, 0]
mean2 = [4, 4]
cov = [[1, 0.5], [0.5, 1]]
gauss1 = np.random.multivariate_normal(mean1, cov, 500)
gauss2 = np.random.multivariate_normal(mean2, cov, 500)
gaussian_data = np.vstack([gauss1, gauss2])

# Banana-Shaped (Non-Gaussian)
theta = np.linspace(0, np.pi, 1000)
x = theta + 0.1 * np.random.randn(1000)
y = np.sin(theta) + 0.1 * np.random.randn(1000)
banana_data = np.stack([x, y], axis=1)
# Normalize Data
scaler_gauss = StandardScaler()
gaussian_data_scaled = scaler_gauss.fit_transform(gaussian_data)
scaler_banana = StandardScaler()
banana_data_scaled = scaler_banana.fit_transform(banana_data)
print("Gaussian data scaled mean:", gaussian_data_scaled.mean(axis=0))
print("Gaussian data scaled std:", gaussian_data_scaled.std(axis=0))
print("Banana data scaled mean:", banana_data_scaled.mean(axis=0))
print("Banana data scaled std:", banana_data_scaled.std(axis=0))

# 2. Create Flow
def create_flow(input_dim=2, hidden_features=128, num_layers=6):
    transforms = []
    masks = [torch.tensor([1, 0]), torch.tensor([0, 1])] * (num_layers // 2)
    for mask in masks:
        transforms.append(ReversePermutation(features=input_dim))
        transforms.append(
            PiecewiseRationalQuadraticCouplingTransform(
                mask=mask,
                transform_net_create_fn=lambda in_features, out_features: ResidualNet(
                    in_features=in_features,
                    out_features=out_features,
                    hidden_features=hidden_features,
                    num_blocks=3,
                    activation=nn.ReLU()
                ),
                tails='linear',
                num_bins=10
            )
        )
    transform = CompositeTransform(transforms)
    base_dist = StandardNormal(shape=[input_dim])
    return Flow(transform, base_dist)

# 3. Train Flow
def train_flow(flow, data, scaler, dataset_name, steps=15000, lr=1e-4):
    optimizer = torch.optim.Adam(flow.parameters(), lr=lr)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3000, gamma=0.5)
    data_tensor = torch.tensor(data, dtype=torch.float32)
    losses = []
    batch_size = 256
    for step in range(steps):
        idx = np.random.choice(len(data), size=batch_size, replace=False)
        batch = data_tensor[idx]
        loss = -flow.log_prob(batch).mean()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        scheduler.step()
        losses.append(loss.item())
        if step % 500 == 0:
            print(f"Step {step} | Loss: {loss.item():.4f} | LR: {scheduler.get_last_lr()[0]:.6f}")
        if step in [5000, 10000]:
            print(f"Evaluating {dataset_name} at step {step}...")
            evaluate_fit(flow, data, scaler, label=f"{dataset_name} (Step {step})")
    plt.figure(figsize=(8, 6))
    plt.plot(losses)
    plt.xlabel("Step")
    plt.ylabel("Loss")
    plt.title(f"Training Loss Curve - {dataset_name}")
    plt.grid(True)
    plt.show()
    return losses

# 4. KL Divergence Function
def kl_divergence_1d(p, q):
    epsilon = 1e-10
    p = np.clip(p, epsilon, 1)
    q = np.clip(q, epsilon, 1)
    return np.sum(p * np.log(p / q))

# 5. Evaluation Function
def evaluate_fit(flow, real_data, scaler, label=""):
    samples = flow.sample(len(real_data)).detach().numpy()
    samples_original = scaler.inverse_transform(samples) if scaler is not None else samples
    real_data_original = scaler.inverse_transform(real_data) if scaler is not None else real_data

    # Scatter Plot
    plt.figure(figsize=(8, 6))
    plt.scatter(real_data_original[:, 0], real_data_original[:, 1], alpha=0.4, label="Real Data")
    plt.scatter(samples_original[:, 0], samples_original[:, 1], alpha=0.4, label="Generated Samples")
    plt.title(f"{label} - Real vs Generated Samples")
    plt.legend()
    plt.grid(True)
    plt.show()

    # KDE and KL Divergence
    kl_scores = []
    for i in range(2):
        plt.figure(figsize=(8, 6))
        p = gaussian_kde(real_data_original[:, i])
        q = gaussian_kde(samples_original[:, i])
        xs = np.linspace(
            min(real_data_original[:, i].min(), samples_original[:, i].min()),
            max(real_data_original[:, i].max(), samples_original[:, i].max()), 1000
        )
        p_vals = p(xs)
        q_vals = q(xs)
        plt.plot(xs, p_vals, label="Real KDE")
        plt.plot(xs, q_vals, label="Generated KDE")
        plt.title(f"{label} - Axis {i}")
        plt.legend()
        plt.grid(True)
        plt.show()

        kl = kl_divergence_1d(p_vals, q_vals)
        kl_scores.append(kl)
        print(f"{label} KL divergence (axis {i}): {kl:.4f}")

    avg_kl = np.mean(kl_scores)
    print(f"{label} Avg KL: {avg_kl:.4f}")
    return avg_kl

# 6. Run Training and Evaluation
print("Training on Gaussian data...")
flow_gauss = create_flow()
train_flow(flow_gauss, gaussian_data_scaled, scaler_gauss, "Gaussian")
evaluate_fit(flow_gauss, gaussian_data_scaled, scaler_gauss, label="Gaussian (Final)")

print("\nTraining on Banana data...")
flow_banana = create_flow()
train_flow(flow_banana, banana_data_scaled, scaler_banana, "Banana")
evaluate_fit(flow_banana, banana_data_scaled, scaler_banana, label="Banana (Final)")
```
